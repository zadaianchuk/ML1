```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

from scipy import stats
from sklearn.linear_model import LinearRegression
```

### 1. Load and clean data
So we need to read the data, name columns and take mean over Videos for each Subject

```python
data=pd.read_csv("./data/EEG data.csv", header=None)

names={0: "Subject ID", 1: "Video ID",2:"Attention",3:"Mediation",4:"Raw",5: "Delta",6:" Theta",7: "Alpha 1",8: "Alpha 2",9: "Beta 1", 10: "Beta 2" ,11: "Gamma 1", 12: "Gamma 2",13: "predefined label", 14: "user-defined label"}
data=data.rename(columns=names)

data=data.groupby(["Subject ID", "Video ID"]).mean()
```

### 2. Outlier Analysis

Let us look at the attention and Mediation columns.

```python
sns.boxplot(data[["Attention","Mediation"]])
```

So there are several outliers. First that we can see, that there are some 0 values. Let's find them. 

```python
data[(data["Attention"]==0)|(data["Mediation"]==0)]
```

These outliers 0 values are common.

I will just delete all the rows above.  

```python
data=data[~((data["Attention"]==0)|(data["Mediation"]==0))]
data[(data["Attention"]==0)|(data["Mediation"]==0)]
```

After we delete some recordings that are obviously spoiled.,we can look once more at our data. 

```python
sns.boxplot(data[["Attention","Mediation"]])
```

Let's delete all the values outside $\pm 3\sigma$

```python
data[(data["Mediation"]>data["Mediation"].quantile(0.9985))|
     (data["Mediation"]<data["Mediation"].quantile(0.0015))|
     (data["Attention"]>data["Attention"].quantile(0.9985))|
     (data["Attention"]<data["Attention"].quantile(0.0015))]
```

```python
data=data[~((data["Mediation"]>data["Mediation"].quantile(0.9985))|
       (data["Mediation"]<data["Mediation"].quantile(0.0015))|
       (data["Attention"]>data["Attention"].quantile(0.9985))|
       (data["Attention"]<data["Attention"].quantile(0.0015)))]
```

```python
sns.boxplot(data[["Attention","Mediation"]])
```

### 3. Predictor distributions and correlations

```python
features =["Delta"," Theta", "Alpha 1", "Alpha 2", "Beta 1", "Beta 2" , "Gamma 1", "Gamma 2"]
X=data[features]
skew=X.skew(axis=0)
# Set up the matplotlib figure
f, axes = plt.subplots(2,4, figsize=(20, 10), sharex=False)
sns.despine(left=True)
for ax,feature in zip(axes.flatten(),features): 
    sns.distplot(data[feature].values, color="b", ax=ax)
    ax.set_title("skewness {0:.2f}".format(skew[feature]))
plt.setp(axes, yticks=[]);
plt.setp(axes, xticks=[]);
```

Let's apply Box-Cox correction to the features where skewness is larger than $\frac{2}{3}$.

```python
print(skew)
for feature in list(skew[np.abs(skew)>2/3].index):
    X.loc[:,feature]=stats.boxcox(X.loc[:,feature].values)[0]
```

Let's check that we correct all skewed features

```python
X.skew(axis=0)
```

```python
f, ax = plt.subplots(figsize=(10, 8))
corr = X.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
```

We can see that slow frequency features correlate with each other.

It is also useful to plot correlation between every pair of features. 

```python
plt.figure(figsize=(20,20))
sns.set(style="ticks", color_codes=True)
sns.pairplot(X)
```

### 3. Basic linear regression

I decided to add linear regression line with 95% confidence interval. 

```python
features = ["Delta"," Theta", "Alpha 1", "Alpha 2", "Beta 1", "Beta 2" , "Gamma 1", "Gamma 2"]
targets = ["Attention", "Mediation"]
f, axes = plt.subplots(8, 2, figsize=(20, 30), sharex=False)
for i,feature in enumerate(features):
    for j,target in enumerate(targets):
            sns.regplot(x=feature, y=target, data=data,ax=axes[i,j])
```

```python
linear_model=LinearRegression(normalize=True)

# I get numply arrays of from DataFrames 
X=data[features].values
y_att=data["Attention"].values
y_med=data["Mediation"].values
```

```python
y=y_att
linear_model.fit(X,y)
y_hat=linear_model.predict(X)
RMSE=(((y-y_hat)**2).sum())**(1/2)
print("RMSE for attention {0:.2f}".format(RMSE))
y=y_med
linear_model.fit(X,y)
y_hat=linear_model.predict(X)
RMSE=(((y-y_hat)**2).sum())**(1/2)
print("RMSE for mediation {0:.2f}".format(RMSE))
```

Let's add some interactions $x_6 \cdot x_0$, $x_6 \cdot x_2$ ,$x_6 \cdot x_4$  and $x_i^2$

```python
X=data[features].values
X=np.c_[X,X**2, X[:,6]*X[:,0],X[:,6]*X[:,2],X[:,6]*X[:,4]]
y=y_att
linear_model.fit(X,y)
y_hat=linear_model.predict(X)
data["Attention fitted"]=y_hat
data["Attention residuals"]=y-y_hat
RMSE=(((y-y_hat)**2).mean())**(1/2)
print("RMSE for attention {0:.2f}".format(RMSE))
y=y_med
linear_model.fit(X,y)
y_hat=linear_model.predict(X)
data["Mediation fitted"]=y_hat
data["Mediation residuals"]= y-y_hat
RMSE=(((y-y_hat)**2).mean())**(1/2)
print("RMSE for mediation {0:.2f}".format(RMSE))
```

```python

sns.lmplot("Attention fitted", "Attention residuals",data=data,fit_reg=False,scatter_kws={"marker": "D","s": 100}) 

sns.lmplot("Mediation fitted", "Mediation residuals",data=data,fit_reg=False,scatter_kws={"marker": "D","s": 100}) 
```

Looks gut! Lets check normality 

```python
import pylab
fig=plt.figure(figsize=(15,5))
fig.add_subplot(121)
stats.probplot(data["Attention residuals"], dist="norm", plot=pylab)
fig.add_subplot(122)
stats.probplot(data["Mediation residuals"], dist="norm", plot=pylab)
pylab.show()
```

### 5. Linear regression for prediction

```python
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.metrics import mean_squared_error
```

```python
X=data[features].values
X=np.c_[X,X**2, X[:,6]*X[:,0],X[:,6]*X[:,2],X[:,6]*X[:,4]]
y=y_att
scores = cross_val_score(linear_model, X, y, cv=5,scoring='neg_mean_squared_error')
RMSE=np.sqrt(-scores).mean()     
print("RMSE for attention {0:.2f}".format(RMSE))
y=y_med
scores = cross_val_score(linear_model, X, y, cv=5,scoring='neg_mean_squared_error')
RMSE=np.sqrt(-scores).mean()  
print("RMSE for mediation {0:.2f}".format(RMSE))
```

The RMSE is higher than it was when we looked at the whole dataset. It indicated that before we overfit to our data. So maybe we need to add regularisation

```python
y=y_att
scores = cross_val_score(linear_model, X_new, y, cv=5,scoring='neg_mean_squared_error')
RMSE=np.sqrt(-scores).mean()     
print("RMSE for attention {0:.2f}".format(RMSE))
y=y_med
scores = cross_val_score(linear_model, X_new, y, cv=5,scoring='neg_mean_squared_error')
RMSE=np.sqrt(-scores).mean()  
print("RMSE for mediation {0:.2f}".format(RMSE))
```
